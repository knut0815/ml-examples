Notebooks are interactive development environment that allows you to write code step-by-step, saving and re-running chunks of code as necessary. Notebooks are best for exploration and explanation, but if you want to re-use your work somewhere else you can export the code as Python script. In Jupyter click `File > Download As > Python (.py)`, and in Colab click `File > Download .py`.

Jupyter used to be called "IPython Notebook" and a lot of documentation refers to it that way still. But now Jupyter supports other languages too (like Lua/Torch with iTorch) and it's more general purpose. A complete introduction to Jupyter is available [here](http://jupyter-notebook.readthedocs.io/en/latest/notebook.html).

Colab is a notebook system developed by Google that usually runs on their free cloud machines. One big advantage of Colab is the ability to use GPU time for free.

In Notebooks, code is divided into cells. Evaluate a cell by selecting it and then clicking `Cell > Run` (Jupyter), clicking the "play" button (Colab), or by typing `Shift + Enter` (Jupyter & Colab). Try it on the next cell.

```
print('Hello world!!')
```

Besides printing out things you tell it to print, notebooks will also print out the value of the last line in a cell.

```
2 + 2
```

All data is stored "behind the scenes" and can be passed between cells. First evaluate the next cell:

```
word = 'little'
```

Then evaluate the next cell.

```
phrase = word + ' by ' + word
print(phrase)
```

If an operation is running for too long, we can cancel it by pressing the "stop" button above (Jupyter) or to the left of the cell (Colab). Try running the following cell then pressing the "stop" button.

```
from time import sleep
for i in range(60):
    sleep(1)
```

Normally the cell will print some error information about the line it stopped on, similar to pressing `Ctrl-C` in a shell.

If your code is really stuck, sometimes you need to restart everything. To do that, press the `Restart Kernel` button (Jupyter) or `Runtime > Restart runtime` (Colab). This will clear any intermediate results you have, including any saved variables and defined functions.

One more really important thing to know about notebooks is that they interface nicely with matplotlib, a plotting library. In Jupyter, matplotlib usually renders in a pop-up window, but we can import matplotlib and use a special line beginning with `%` (called a "[magic](http://nbviewer.ipython.org/github/ipython/ipython/blob/1.x/examples/notebooks/Cell%20Magics.ipynb)") to mak sure it renders in the notebook.

```
%matplotlib inline
from matplotlib import pyplot as plt
from math import sin
plt.plot([sin(x/10.0) for x in range(100)])
```

Normally we would write `plt.show()` to show the plot, but `plt.plot()` returns the plot as a result, and notebooks render the last line automatically.

It's also possible to render higher output for retina displays with matplotlib.

```
%config InlineBackend.figure_format = 'retina'
plt.plot([sin(x/10.0) for x in range(100)])
```
